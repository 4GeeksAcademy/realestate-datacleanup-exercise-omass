#### Exercise 09. Is the average of "Valdemorillo" and "Galapagar" prices the same? (★★☆)

Print the both average prices and then write a conclusion about them

```python
# TODO

import pandas as pd

df = pd.DataFrame(data_frame)

name1 = 'Valdemorillo'
name2 = 'Galapagar'

data_names = df[df['price'].isin([name1,name2])]

average_for_name1 = data_names.loc[data_names['level5'] == name1, 'level5'].mean()
average_for_name2 = data_names.loc[data_names['level5'] == name2, 'level5'].mean()


if average_for_name1 > average_for_name2:
    print(f'{name1} average price {average_for_name1} is more than {name2}')
elif average_for_name1 < average_for_name2:
    print(f'{name1} average price {average_for_name1} is less than {name2}')
else:
    print(f'{name1} average price {average_for_name1} is equal to {name2}')
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO
import matplotlib.pyplot as plt
import pandas as pd


south_belt_ds = df[df["level5"].notna()]  
populations = south_belt_ds["level5"].unique()[:4]

fig, axes = plt.subplots(nrows=len(populations), ncols=1, figsize=(8, 2 * len(populations)))

for i, population in enumerate(populations):
    population_data = south_belt_ds[south_belt_ds["level5"] == population]
    
    
    axes[i].scatter(population_data.index, population_data["price"], label=population)
    axes[i].set_title(f"Prices for {population}")
    axes[i].set_xlabel("Index")
    axes[i].set_ylabel("Price")
    axes[i].legend()

plt.tight_layout()
plt.show()
```
